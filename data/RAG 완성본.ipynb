````python
import os
from langchain_community.document_loaders import CSVLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_google_genai import GoogleGenerativeAIEmbeddings, ChatGoogleGenerativeAI
from langchain_community.vectorstores import FAISS
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
import json

# API 키 설정
os.environ["GOOGLE_API_KEY"] = st.secrets["GEMINI"]["API_KEY"]

# 1. CSV 데이터 로드
csv_path = "/content/drive/MyDrive/Colab Notebooks/book_data.csv"
loader = CSVLoader(
    file_path=csv_path,
    csv_args={'delimiter': ',', 'quotechar': '"'},
    encoding='utf-8-sig',
    metadata_columns=['BOOK_TITLE_NM', 'AUTHR_NM', 'BOOK_IMAGE_NM']
)
documents = loader.load()

# 2. 텍스트 분할 (chunk_size=300, overlap=50)
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=300,
    chunk_overlap=50,
    length_function=len
)
splits = text_splitter.split_documents(documents)

# 3. 임베딩 모델 설정 (gemini-embedding-001)
embeddings = GoogleGenerativeAIEmbeddings(model="text-embedding-004")

# 4. 벡터 스토어 생성
vectorstore = FAISS.from_documents(splits, embeddings)

# 5. Retriever 설정 (k=20)
retriever = vectorstore.as_retriever(search_kwargs={"k": 20})

# 6. LLM 모델 설정 (gemini-2.5-flash, temperature=0.7)
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=0.7
)

# 7. 프롬프트 템플릿 작성
prompt_template = """당신은 사용자의 요청에 맞춰 책을 추천하는 '전문 도서 큐레이터'입니다.

**임무:**
사용자의 요청(주제, 장르, 키워드)을 바탕으로, 아래 [검색된 책 목록] 중에서 가장 관련성이 높은 책 3~5권을 추천해야 합니다.

**엄격한 규칙:**
1. **JSON 출력:** 응답은 반드시 아래 JSON 형식이어야 합니다.
2. **컨텍스트 기반:** 추천은 오직 [검색된 책 목록]에 있는 정보만을 기반으로 해야 합니다. 절대로 목록에 없는 책을 지어내거나 외부 지식을 사용하지 마세요.
3. **정확한 정보 추출:** 각 책의 '제목(BOOK_TITLE_NM)', '저자(AUTHR_NM)', '이미지URL(BOOK_IMAGE_NM)' 필드의 값은 검색된 문서에서 정확히 가져와야 합니다.
4. **추천 이유 생성:** 'reason' 필드에는 검색된 책의 '책 소개' 내용을 바탕으로, 왜 이 책이 사용자의 요청과 잘 맞는지 1~2문장으로 설득력 있게 작성해야 합니다.
5. **적절한 개수:** 3권에서 5권 사이로 추천합니다. 만약 관련 있는 책이 아예 없다면, 'recommendations' 리스트를 비워두고 'message'에 "요청하신 주제와 관련된 책을 찾지 못했습니다."라고 응답하세요.
6. **메시지 작성:** 'message' 필드에는 "요청하신 주제로 {{추천 책 개수}}권의 책을 찾았습니다." 또는 실패 메시지를 작성하세요.

[검색된 책 목록]:
{context}

[사용자 요청]:
{question}

**출력 형식 (JSON만 출력하세요):**
{{
    "message": "요청하신 주제로 N권의 책을 찾았습니다.",
    "recommendations": [
        {{
            "title": "BOOK_TITLE_NM 값",
            "author": "AUTHR_NM 값",
            "image_url": "BOOK_IMAGE_NM 값",
            "reason": "이 책을 추천하는 이유 (1~2문장)"
        }}
    ]
}}

JSON 형식만 출력하고, 다른 설명이나 마크다운은 추가하지 마세요.
"""

PROMPT = PromptTemplate(
    template=prompt_template,
    input_variables=["context", "question"]
)

# 8. 도서 추천 함수
def recommend_books(user_query):
    """
    사용자 질문을 받아 도서를 추천하는 함수

    Args:
        user_query (str): 사용자의 키워드, 제목, 주제, 장르 등

    Returns:
        dict: JSON 형태의 추천 도서 목록
    """
    # Retriever로 관련 문서 검색
    retrieved_docs = retriever.get_relevant_documents(user_query)

    # 컨텍스트 포맷팅 (메타데이터 명시적으로 포함)
    context_parts = []
    for i, doc in enumerate(retrieved_docs, 1):
        metadata = doc.metadata
        context_parts.append(
            f"[책 {i}]\n"
            f"제목(BOOK_TITLE_NM): {metadata.get('BOOK_TITLE_NM', 'N/A')}\n"
            f"저자(AUTHR_NM): {metadata.get('AUTHR_NM', 'N/A')}\n"
            f"이미지URL(BOOK_IMAGE_NM): {metadata.get('BOOK_IMAGE_NM', 'N/A')}\n"
            f"책 소개: {doc.page_content}\n"
        )

    context = "\n".join(context_parts)

    # LLM 체인 실행
    chain = LLMChain(llm=llm, prompt=PROMPT)
    result = chain.invoke({"context": context, "question": user_query})

    try:
        # JSON 파싱
        response_text = result['text']
        # JSON 부분만 추출 (마크다운 코드 블록 제거)
        if "```json" in response_text:
            json_str = response_text.split("```json")[1].split("```")[0].strip()
        elif "```" in response_text:
            json_str = response_text.split("```")[1].split("```")[0].strip()
        else:
            json_str = response_text.strip()

        recommendations = json.loads(json_str)
        return recommendations
    except json.JSONDecodeError as e:
        print(f"JSON 파싱 오류: {e}")
        print(f"원본 응답: {result['text']}")
        return {"error": "응답 파싱 실패", "raw_response": result['text']}

# 9. 사용 예시
if __name__ == "__main__":
  pass
````
